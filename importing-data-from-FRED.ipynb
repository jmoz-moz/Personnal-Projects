```python
from pandas_datareader.data import DataReader
from datetime import date
import matplotlib.pyplot as plt

series_code = 'DGS10'
data_source = 'fred'
start = date(1962, 1, 1)
data= DataReader(series_code, data_source, start)

data.plot(title='10-year Treasury')
plt.show()
```

```python
# Set start date
start = date(1968, 1, 1)

# Set series code
series = 'GOLDAMGBD228NLBM'
data_source='fred'
# Import the data
gold_price = DataReader(series, data_source, start=start)

# Inspect the price of gold
gold_price.info()

# Plot the price of gold
gold_price.plot(title='Gold Price')

# Show the plot
plt.show()
```

```python
# Set the start date
start = date(2008, 1, 1)

# Set the series codes
series = ['BAMLHYH0A0HYM2TRIV', 'SP500']
data_source= 'fred'
# Import the data
data = DataReader(series, data_source, start)

# Plot the results
data.plot(title='Performance Comparison', subplots=True)

# Show the plot
plt.show()
```
