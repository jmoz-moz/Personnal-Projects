## Converting datatypes: date, store, and holiday


```python
df.columns = ['store', 'date', 'weekly_sales', 'holiday', 'temp', 'fuel_price', 'cpi', 'unemployment']
#changing column names to lowercase
```

```python
df.date = pd.to_datetime(df.date)
#converting date column from objecto to date datatype
```

```python
df['year'], df['month'], df['day'] = df['date'].dt.year, df['date'].dt.month, df['date'].dt.day
#adding three new columns for year, month, and day of each row
```

## Comverting the store column into category since all those values are unique and pertaining to its store number. 

## As well as the holiday, Boolean.

```python
df.store = df.store.astype('category')
```

```python
df.holiday = df.holiday.astype('category')
```

```python
df.isnull().sum()
df.duplicated().sum()

#no duplicates and no null values
```
